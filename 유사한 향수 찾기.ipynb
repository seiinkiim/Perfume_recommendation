# 코사인 유사도 계산

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
X_simi = np.zeros((len(perfume_1), len(perfume_1)))
```

```python
for i in range(len(X_simi)):
    for j in range(len(X_simi)):
        X_simi[i][j] = cos_sim(X[i], X[j])
```

### 508개의 향수, [i, j]성분은 i번째 향수와 j번째 향수의 코사인 유사도를 의미1!!!

```python
X_simi
```

```python
# 대각선 성분은 모두 0으로!!
for i in range(len(X_simi)):
    X_simi[i][i] = 0
```

### 대략적으로 코사인 유사도를 확인했을 때, 0.1~0.3은 너무 많은 유사한 향수가 나오고, 0.5 이상으로 하면 대부분의 향수가 유사 향수가 검색되지 않습니다.

```python
print("more than 0.1 : ", sum(X_simi>=0.1).mean())
print("more than 0.2 : ",sum(X_simi>=0.2).mean())
print("more than 0.3 : ",sum(X_simi>=0.3).mean())
print("more than 0.4 : ",sum(X_simi>=0.4).mean())
print("more than 0.5 : ",sum(X_simi>=0.5).mean())
print("more than 0.6 : ",sum(X_simi>=0.6).mean())
print("more than 0.7 : ",sum(X_simi>=0.7).mean())
print("more than 0.8 : ",sum(X_simi>=0.8).mean())
print("more than 0.9 : ",sum(X_simi>=0.9).mean())
print("more than 1.0 : ",sum(X_simi>=1.0).mean())
```

### 5개 이상 검색되는 향수는 상위 5개를, 1-4개로 검색되는 향수는 상위 1-4개를 저장해줍니다.

```python
def find_indexes(dictionary, target_value):
    indexes = []
    for key, value in dictionary.items():
        if value == target_value:
            indexes.append(key)
    return indexes



indexes_1 = find_indexes(simi_dict, 1)
indexes_2 = find_indexes(simi_dict, 2)
indexes_3 = find_indexes(simi_dict, 3)
indexes_4 = find_indexes(simi_dict, 4)
indexes_5 = find_indexes(simi_dict, 5)

print(indexes_1)
print(indexes_2)
print(indexes_3)
print(indexes_4)
```

### 상위 5개의 값의 인덱스를 추출하는 코드

```python
def get_top_indices(array, num_indices):
    # 배열에서 상위 num_indices 개의 값을 추출하는 함수
    top_indices = np.argsort(array)[-num_indices:][::-1]
    return top_indices

# 예시 배열
arr = np.array(X_simi[0])

# 상위 5개 값의 인덱스 추출
top_indices = get_top_indices(arr, 5)

print(top_indices)
```
